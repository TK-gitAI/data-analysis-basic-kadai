```python
import matplotlib.pyplot as plt
import numpy as np

nrows = 2
ncols = 3
fig, axes = plt.subplots(nrows, ncols)

x0 = [0,1,2,3]
y0 = [1, 2, -5, 2]
axes[0,0].plot(x0,y0)

x0 = np.linspace(0,10,100)
y1 = 2 + 2 *np.sin(2 * x0)
axes[0,1].plot(x0,y1,linewidth=2)
axes[0,1].set(xlim=(0,10),xticks=np.arange(0,10),ylim=(-1,5),yticks=np.arange(-1,6))

x0 = np.random .normal(15,5,2000)
axes[0,2].hist(x0)

np.random.seed(3)
x1 = 4 + np.random.normal(0,2,60)
y0 = 4 + np.random.normal(0,2,len(x1))
axes[1,0].scatter(x1,y0)

plt.show()
```
